LIMPIEZA DE DATOS DEL DATASET HR-EMPLOYEE-ATTRITION PARA SU POSTERIOR VISUALIZACION EN POWER BI


```python
# Importamos las librerias necesarias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Abrimos nuestro dataset
df = pd.read_csv('HR-Employee-Attrition.csv')
df
```

```python
#Eliminamos columnas que no son necesarias
df1 = df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'StandardHours'])
df1
```

```python
#Buscaremos todos los valores nulos
valores_nulos = df1.isnull().sum()
#print(valores_nulos)
#Vamos a mostrar la informacion del dataset
#print(df1.info())
```

```python
# Coversion de variables muestra la rotacion de empleados
df1['Attrition'] = df1['Attrition'].map({'Yes': 1, 'No': 0})
print("\n-- 'Attrition ' convertido a numerico (0=No, 1=Yes) ---")
print(df1['Attrition'].value_counts())
```

```python
#Mapeo de variables numericas codificadas a etiquitas legibles para mejorar la legibilidad en Power Bi
mapeo_educacion = {
    1: '1 - Below College', 2: '2 - College', 3: '3 - Bachelor', 4: '4 - Master', 5: '5 - Doctorate'
}
df1['Education'] = df1['Education'].map(mapeo_educacion)

satisfacion_mapping = {
    1: '1 - Low', 2: '2 - Medium', 3: '3 - High', 4: '4 - Very High'
}
df1['EnvironmentSatisfaction'] = df1['EnvironmentSatisfaction'].map(satisfacion_mapping)
df1['JobSatisfaction'] = df1['JobSatisfaction'].map(satisfacion_mapping)
df1['RelationshipSatisfaction'] = df1['RelationshipSatisfaction'].map(satisfacion_mapping)
df1['JobInvolvement'] = df1['JobInvolvement'].map(satisfacion_mapping)

worklifebalance_mapping = {
    1: '1- Bad', 2: '2- Good', 3:'3 - Better', 4: '4 - Best'
}
df1['WorkLifeBalance'] = df1['WorkLifeBalance'].map(worklifebalance_mapping)

perfomance_rating_mapping = {
    3: '3- Excellent', 4: '4 - Outstanding'
}
df1['PerformanceRating'] = df1['PerformanceRating'].map(perfomance_rating_mapping)

print("\n--- 'Culumnas mapeadas a etiquetas legibles(primeras filas)---'")
print(df1[['Education', 'EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction',
           'JobInvolvement', 'WorkLifeBalance', 'PerformanceRating']].head())

print("\n --- Tipos de datos despues de la estandarizacion ---")
print(df1.info())
```

```python
#vamos agrupar las edades para analizar la rotacion por segmentos demograficos
df1['Age_group'] = pd.cut(df1['Age'], 
                          bins=[18, 25, 35, 45, 55, 65],
                          labels=['18-24', '25-34', '35-44', '45-54', '55-64'],
                          right=False,
                          include_lowest=True)
print("\n--- Nueva caracteristica: 'Age_group'---")
print(df1['Age_group'].value_counts().sort_index())

#Vamos a retiro mensula ppr cada año en la empresa
df1['montly_income_per_year_at_company'] = df1['MonthlyIncome'] / (df1['YearsAtCompany'] + 1)
print("\n--- Nueva caracteristica: 'montly_income_per_year_at_company'---")
print(df1[['MonthlyIncome', 'YearsAtCompany', 'montly_income_per_year_at_company']].head())

#Indicadores de experiencia Interna vs externa
#Nos maostrare que empleados tienen mas experiencia externa que interna en la empresa
df1['Experience_Internal'] = df1['YearsAtCompany'] / (df1['TotalWorkingYears'])
print("\n--- Nueva caracteristica: 'Experience_Internal'---")
print(df1[['YearsAtCompany', 'TotalWorkingYears', 'Experience_Internal']].head())

#Verifciamos la antiguedad de cada empleado
#Puede indicar estancamiento si 'years_in_current_role' es muy cercana a 'years_at_company'
df1['stagnation_in_role'] = np.where(
    df1['YearsAtCompany'] > 0,
    df1['YearsInCurrentRole'] / df1['YearsAtCompany'],
    0
)
print("\n--- Nueva caracteristica: 'stagnation_in_role'---")
print(df1[['YearsAtCompany', 'YearsInCurrentRole', 'stagnation_in_role']].head())

# Combinancion de columnas de JobRole y Department para tener un contexto mas expecifico
df1['JobRole_Department'] = df1['JobRole'] + '_' + df1['Department']
print("\n--- Nueva caracteristica: 'JobRole_Department'---")
print(df1[['JobRole', 'Department', 'JobRole_Department']].head())

```

```python
# Vamos a revisar los datos finales
print("\n--- Tipos de datos finales despues de la tramsformaciones ---")
print(df1.info())

#Vamos a revisar los datos estadisticos en la neuvas columnas numericas
print("\n--- Descripciones estadisticas de las columnas numericas ---")
print(df1[['montly_income_per_year_at_company', 'Experience_Internal', 'stagnation_in_role']].describe())

#Vamos a exporta el nuevo archivo para ser utilizado en Power BI
nuevo_archivo_procesado = 'HR_Employee_Attrition_Enhanced_for_PowerBI.csv'
df1.to_csv(nuevo_archivo_procesado, index=False)
print(f"\n --- Archivo exportado correctamente: {nuevo_archivo_procesado} ---")
```
